# Cebu

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline

filename = "./data/itm-final-project-cebu.csv"
df = pd.read_csv(filename)

df['hour'] = pd.to_datetime(df['time']).dt.hour #https://intellipaat.com/community/32903/pandas-convert-strings-to-time-without-date
grouped_df = df.groupby(['hour'])['total_reacts'].sum()
grouped_df = grouped_df.to_frame()
grouped_df.reset_index(inplace=True)

# for bar graph
fig, ax = plt.subplots(figsize = (12,7))

ax.bar(height=grouped_df['total_reacts'] , x=grouped_df['hour'], color = 'DarkBlue')
ax.set_xticks(list(grouped_df['hour']))
ax.set_xlabel('Hour (24H Format)')
ax.set_ylabel('Total Reacts')
ax.set_title('Total Reacts per Hour in Cebu')
plt.show()
```

```python
# create scatter plot for total reacts VS. posting time
fig, ax = plt.subplots(figsize = (12,7))

ax.scatter(grouped_df['hour'], grouped_df['total_reacts'])
ax.set_xlabel('Hour')
ax.set_ylabel('Total Reacts')
ax.set_title('Total Reacts VS Hour')
```

```python
# Pearson's Correlation Test - posting time vs total reactions
grouped_df.corr()
```
